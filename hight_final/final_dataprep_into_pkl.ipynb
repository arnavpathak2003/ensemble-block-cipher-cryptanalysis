```python
import pandas as pd
from pathlib import Path
```

```python
DIR = Path("dataset_pkl_round_9")
DIR.mkdir(exist_ok=True)
```

```python
for i in range(1, 65, 1):
    df_hight_flipped = pd.read_pickle(
        f"HIGHT_Output_Flipped_Bits_round_9/HIGHT_9_round_delta-{i}.pkl"
    )
    df_hight_diff = pd.read_pickle("HIGHT_Output_Diff_Bits/HIGHT_9_round.pkl")

    df_combined = pd.concat([df_hight_flipped, df_hight_diff], ignore_index=True)
    df_combined.to_pickle(f"dataset_pkl_round_9/HIGHT_9_round_delta-{i}_combined.pkl")
```
