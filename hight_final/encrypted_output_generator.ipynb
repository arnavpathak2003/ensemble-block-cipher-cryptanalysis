```python
from utils.hight import hight_encryption_binary
import pandas as pd
import numpy as np
from pathlib import Path
import random
```

```python
random.seed(42)
mk = random.getrandbits(128)
mk
```

```python
for round in range(24, 33, 1):
    PICKLE_DIR = Path(f"HIGHT_Output_Flipped_Bits_round_{round}")
    PICKLE_DIR.mkdir(parents=True, exist_ok=True)
    for delta in range(64, 0, -1):
        all_rows_data = []
        df = pd.read_pickle(f"flipped_bits_raw/bit_flipped_delta-{delta}.pkl")
        P1 = df["P1"].values
        P2 = df["P2"].values
        for i in range(100096):
            plain_text_1 = P1[i]
            plain_text_2 = P2[i]
            C1 = hight_encryption_binary(plain_text_1, mk, round)
            C2 = hight_encryption_binary(plain_text_2, mk, round)
            C = result = np.concatenate([C1, C2])
            all_rows_data.append({"Encrypted_data": C, "Class": "flipped_bits"})
        df = pd.DataFrame(all_rows_data)
        df.to_pickle(
            f"HIGHT_Output_Flipped_Bits_round_{round}/HIGHT_{round}_round_delta-{delta}.pkl"
        )
        print(f"Saved {65 - delta}/64 files....")
```
