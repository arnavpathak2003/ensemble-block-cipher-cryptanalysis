```python
import numpy as np
import pandas as pd

np.random.seed(42)
```

```python
from pathlib import Path

DIR = Path("flipped_bits_raw")
DIR.mkdir(parents=True, exist_ok=True)
```

```python
import numpy as np
import pandas as pd

arr = np.random.randint(0, 2, size=(100096, 64), dtype=np.uint8)


for i in range(1, 65):
    bit_position = 64 - i

    arr_p2 = arr.copy()
    arr_p2[:, bit_position] = 1 - arr_p2[:, bit_position]

    # Create DataFrame
    df = pd.DataFrame(
        {
            "P1": list(arr),
            "P2": list(arr_p2),
        }
    )

    # Save with the naming convention
    filename = DIR / f"bit_flipped_delta-{i}.pkl"
    df.to_pickle(filename)
    print(f"Saved {filename} (flipped bit position {bit_position})")
```
